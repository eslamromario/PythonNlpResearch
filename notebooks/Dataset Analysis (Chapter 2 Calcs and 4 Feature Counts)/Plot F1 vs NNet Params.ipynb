```python
%matplotlib inline
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

FONT_SIZE = 15

STYLE = "whitegrid"
sns.set_style(STYLE)
sns.set_style(STYLE, {"xtick.major.size": 1})
```

```python
#Coral Bleaching
pairs1 = [
    (0.680, 1526799),
    (0.673, 552571),
    (0.650, 738811),
    (0.636, 355195),
    (0.635, 237115),
    (0.634, 286651)
]
f1s, num_params = list(zip(*pairs1))
fig = plt.figure(figsize=(12,6.5))

x = num_params
y = f1s

ax1 = fig.add_subplot(111)
ax1.scatter(x,y)

#ax1.set_xlim([1, 6])
#ax1.set_ylim([0.695, 0.77])
ax1.set_xlabel('Number of Parameters', fontsize=FONT_SIZE)
ax1.set_ylabel('Micro F1 Score', fontsize=FONT_SIZE)

sns.despine()
plt.show()
```
