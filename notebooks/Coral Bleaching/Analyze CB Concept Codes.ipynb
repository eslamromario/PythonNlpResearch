```python
import nltk
from nltk.corpus import wordnet as wn

def strip_non_ascii(text):
    return ''.join(i for i in text if ord(i)<128)

def convert2wn_pos(pos):
    if pos.startswith("NN"):
        return wn.NOUN
    elif pos.startswith("VB"):
        return wn.VERB
    elif pos.startswith("JJ"):
        return wn.ADJ
    elif pos.startswith("RB"):
        return wn.ADV
    else:
        return None

memoized = defaultdict(list)
def __get_words_from_synset__(word, pos):
    synsets = wn.synsets(word, pos=pos)
    if len(synsets) == 0:
        return [word]
    
    words = set()
    for syn in synsets:
        for lemma in syn.lemmas():
            words.add(lemma.name())
    return list(words)

def get_word_synonyms(word, pos):
    pair =  (word, pos)
    if pair in memoized:
        return memoized[pair]
    
    words = __get_words_from_synset__(word, pos)
    memoized[pair] = words
    return words
    
def get_synonyms(word, pos):
    if pos is None:
        return [word]
    wn_pos = convert2wn_pos(pos)
    if not wn_pos:
        return [word]
    return get_word_synonyms(word, wn_pos)
```

```python
syn_map2 = {}
mapped = set()
for sentence in all_sentences:
    tags = nltk.pos_tag(map(strip_non_ascii, sentence))
    for wd, tag in tags:
        pair = (wd, tag)
        if pair in mapped:
            continue
        synonyms = [(s,tag) for s in get_synonyms(wd, tag) if s in wd_sent_freq]
        if len(synonyms) >= 1:
            matches = []
            for spair in synonyms:
                if spair in syn_map2:
                    matches.append(syn_map2[spair])
            if len(matches) == 0:
                synset = set(synonyms)
                synset.add(pair)
                # Won't this potentially overwrite existing maps?
                # >> No see code above. It creates a new set that includes the existing mappings and extends it
                for p in synset:
                    syn_map2[p] = synset                
            elif len(matches) == 1:
                matches[0].add(pair)
                syn_map2[pair] = matches[0]
            else:
                #merge existing synonym lists
                new_synset = set()
                for m in matches:
                    new_synset.update(m)
                #update mapping to map to new larger set
                for s in new_synset:
                    syn_map2[s] = new_synset
            mapped.add(pair)
```
